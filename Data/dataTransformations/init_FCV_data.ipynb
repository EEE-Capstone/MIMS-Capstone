```python
import pandas as pd
import numpy as np
```

```python
## There are only three FCVs to consider: Toyota Mirai, Honda Clarity, Hyundai Nexo from 2014-present
column_names = ['Model', 'Make', 'Year', 'atvType', 'cylinders', 'displ', 'trany',
       'VehicleID', 'AdjWeight', 'comb08', 'combinedUF', 'combE', 'highway08',
       'rangeHwyA', 'RangeHwy', 'Range', 'city08', 'UCity', 'UHighway']
## Honda Clarity data
Honda_FCX_Clarity = ['FCX_Clarity', 'Honda', 2008, 'FCV', 0, 0.0, 'Automatic (A1)', 
                    '1', 3500, 59, 0, 0, 60, 0, 231, 0, 58, 0, 0]
Honda_Clarity_FCV = ['Clarity_FCV', 'Honda', 2017, 'FCV', 0, 0.0, 'Automatic (A1)', 
                    '1', 4000, 68, 0, 0, 67, 0, 366, 0, 69, 0, 0]
## Toyota Mirai Data
Toyota_Mirai_Old = ['Mirai XLE', 'Toyota', 2016, 'FCV', 0, 0.0, 'Automatic (A1)', 
                    '1', 4000, 59, 0, 0, 60, 0, 312, 0, 58, 0, 0]
Toyota_Mirai_New = ['Mirai XLE', 'Toyota', 2021, 'FCV', 0, 0.0, 'Automatic (A1)', ## xle (long range version) 
                    '1', 4250, 74, 0, 0, 76, 0, 402, 0, 71, 0, 0]
## Hyundai Nexo data
Hyundai_Nexo = ['Nexo Blue', 'Hyundai', 2019, 'FCV', 0, 0.0, 'Automatic (A1)', ## Blue (long range version) 
                    '1', 4000, 61, 0, 0, 65, 0, 380, 0, 58, 0, 0]

## create arrays for each model for correct model years
Honda_FCX_Clarity_arr = np.tile(np.array(Honda_FCX_Clarity), (7, 1))
Honda_FCX_Clarity_df = pd.DataFrame(Honda_FCX_Clarity_arr, columns = column_names)
Honda_FCX_Clarity_df['Year'] = np.arange(2008,2015)

Honda_Clarity_FCV_arr = np.tile(np.array(Honda_Clarity_FCV), (3, 1))
Honda_Clarity_FCV_df = pd.DataFrame(Honda_Clarity_FCV_arr, columns = column_names)
Honda_Clarity_FCV_df['Year'] = np.arange(2017,2020)

Toyota_Mirai_Old_arr = np.tile(np.array(Toyota_Mirai_Old), (5, 1))
Toyota_Mirai_Old_df = pd.DataFrame(Toyota_Mirai_Old_arr, columns = column_names)
Toyota_Mirai_Old_df['Year'] = np.arange(2016,2021)

Toyota_Mirai_New_arr = np.tile(np.array(Toyota_Mirai_New), (3, 1))
Toyota_Mirai_New_df = pd.DataFrame(Toyota_Mirai_New_arr, columns = column_names)
Toyota_Mirai_New_df['Year'] = np.arange(2021,2024)

Hyundai_Nexo_arr = np.tile(np.array(Hyundai_Nexo), (5, 1))
Hyundai_Nexo_df = pd.DataFrame(Hyundai_Nexo_arr, columns = column_names)
Hyundai_Nexo_df['Year'] = np.arange(2019,2024)
```

```python
## concatenate dfs
Combined_FCVs = pd.concat([Honda_FCX_Clarity_df, Honda_Clarity_FCV_df, Toyota_Mirai_Old_df,
                          Toyota_Mirai_New_df, Hyundai_Nexo_df])
Combined_FCVs = Combined_FCVs.reset_index(drop = True)
Combined_FCVs['VehicleID'] = Combined_FCVs.index
```
