```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
emissionsjoin = pd.read_csv("/Users/josheverts/MIMS-Capstone/Data/emissions_vehicle_join_adjweights_2000-2023.csv")
```

# Calculating values for use in tool validation (in tool comparison document)

```python
## functions from backend used below
def est_battery_capacity_from_vehicle(Vehicle):
    '''return battery capacity in kwh'''
    kwh_size = 0
    if Vehicle.atvType == 'EV':
        kwh_size = Vehicle.Range*(1/Vehicle.highway08)*33.7 ## use Range since all EVs have it
    elif Vehicle.atvType == 'PHEV':
        kwh_size = Vehicle.rangeHwyA*(Vehicle.combE/100) ## range*(kWh/100 miles) = kWh estimated
#     print(kwh_size)
    return kwh_size
def get_fuel_carbon():
    ## future option: 
    #     np.sum(total_fuel*(np.array(fuel_import_region_weights)*np.array(upstream_ghg_intensities)))
    ##
    carbon_per_MJ = 90.5432 ## gCO2e/MJ from GREET WTW calculator US average
    energy_per_gallon = 118227 ## BTU/gallon averaged from https://afdc.energy.gov/files/u/publication/fuel_comparison_chart.pdf
    ## now convert from carbon per MJ to carbon per gallon (.001055 btu per MJ)
    return carbon_per_MJ*energy_per_gallon*.001055  ## from GREET WTW calculator, GREET1 model with US fuel mix
```

```python
## averages to compare to EPA estimates for 2021, divided by lifetime mileage
avg_2021_usage = np.mean(emissionsjoin[(emissionsjoin['Year'] == 2021)]['usage']/178000)
avg_electric_usage = np.mean(emissionsjoin[(emissionsjoin['Year'] == 2021) & ((emissionsjoin['atvType'] == 'EV') | (emissionsjoin['atvType'] == 'PHEV'))]['usage']/178000)
print("Usage 2021:", avg_2021_usage)
print("Usage EV 2021", avg_electric_usage)
```

```python
## what is our adjustment for upstream fuel emissions?
fuel_carbon_diff = (get_fuel_carbon() - 8887) / 8887
print("Upstream Fuel % Difference From Baseline: ", fuel_carbon_diff*100)
```

```python
## lifetime estimates to compare to papers
## Battery estimates
## get avg battery capacity in the dataset
batt_capacities = []
for i in range(len(emissionsjoin)):
    Vehicle = emissionsjoin.iloc[i]
    batt_cap = est_battery_capacity_from_vehicle(Vehicle) 
    if batt_cap > 0: ## phev/ev only
        batt_capacities.append(batt_cap)
battery_emissions = emissionsjoin[emissionsjoin['atvType'] == 'EV']['battery_assembly'] + emissionsjoin[emissionsjoin['atvType'] == 'EV']['battery_manufacture']
print("Mean Battery Capacity", np.mean(batt_capacities))
## emissions/kwh battery capacity
print("Mean emissions/kWH battery (kgCO2e/kWh): ", np.mean(battery_emissions)/(1000*np.mean(batt_capacities)))
```

```python
## total embodied vehicle emissions estimates
total_embodied = emissionsjoin[emissionsjoin['atvType'] == 'ICE']['total_embodied']
print("mean total embodied for ICE (kgCO2e):", np.mean(total_embodied)/1000)
## for mid-size vehicles (to compare to papers)
total_embodied_midsize = emissionsjoin[(emissionsjoin['atvType'] == 'ICE') & ((emissionsjoin['AdjWeight'] < 4000) & (emissionsjoin['AdjWeight'] > 3000))]['total_embodied']
print("mean total embodied for midsize ICE (kgCO2e):", np.mean(total_embodied_midsize)/1000)
## average ICE weight
print("avg ICE weight (kg):", np.mean(emissionsjoin[(emissionsjoin['atvType'] == 'ICE')]['AdjWeight']) / 2.2)
```

```python
## some visual checks of the data
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (13,8))
plt.title('Normalized Emissions Distributions by Drivetrain Type', size = 20)
plt.hist(emissionsjoin[emissionsjoin['atvType'] == 'ICE']['usage']/178000, density = True, label = 'ICE', bins = 20, alpha = 0.4)
plt.hist(emissionsjoin[emissionsjoin['atvType'] == 'EV']['usage']/178000,density = True, label = 'EV', bins = 5, alpha = 1)
plt.hist(emissionsjoin[emissionsjoin['atvType'] == 'PHEV']['usage']/178000,density = True, label = 'PHEV', bins = 5, alpha = 0.4)
plt.hist(emissionsjoin[emissionsjoin['atvType'] == 'HEV']['usage']/178000,density = True, label = 'HEV', bins = 10, alpha = 0.4)
# plt.hist(emissionsjoin[emissionsjoin['atvType'] == 'FCV']['total'],density = True, label = 'HEV', bins = 10, alpha = 0.4)
plt.xlabel('Emissions (grams CO2e)/mi', size = 15)
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize = (13,8))
plt.title('Battery Capacities in Dataset', size = 20)
plt.hist(batt_capacities)
plt.show()
```

```python
## example visualization for tool output
## pick an ICE
Camry = emissionsjoin[emissionsjoin['Model'] == 'Camry']
## pick an EV of similar size
EQS450 = emissionsjoin[emissionsjoin['Model'] == 'EQS 450 4Matic']
import matplotlib.pyplot as plt
x = ['Camry (ICE)', 'EQS450 4-Matic (EV)']
negative_data = [-1*Camry.total_embodied.iloc[0]/1000, -1*EQS450.total_embodied.iloc[0]/1000]
positive_data = [Camry.usage.iloc[0]/1000, EQS450.usage.iloc[0]/1000]

fig = plt.figure(figsize = (13,8))
ax = plt.subplot(111)
ax.barh(x, negative_data, color='darkblue', label = 'embodied')
ax.barh(x, positive_data, color='lightblue', label = 'usage')
plt.legend()
plt.xlabel('Emissions (kgCO2e)', size = 20)
plt.ylabel('Vehicle', size = 20)
```
